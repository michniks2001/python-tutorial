## The `input()` Function

Part of the reason programming has an appeal is the interactivity that programs provide. In order for a user to interact with a program, we have to be able to take input from a user, in order to provide a different results each time a program is executed.

In order to take an input from a user, we must ask a user for something using Python's built in `input()` function, like so.

```python
input('Enter something: ')
```

Of course, just this line of code will do nothing other than allow you to input something onto the console. The real fun can begin when you store what you input into a variable.

```python
name = input('Enter your name: ')
print(f'Hello, {name}!')
```

In that simple example, we were able to create a program that tells the user to input their name and the program responds to the user with a "Hello".

## Type Casting

Now the `input()` function is cool and all, but there is something that needs to be stated about the `input()` function and that is that it stores strings regardless of what you enter. We can use the `type()` function to check the type of a variable.

```python
value = input('Enter a number: ')
print(type(value))
```

Even though your input was a numeric value, the type of your input remains a string.

How do we fix that? With type casting of course! Simply, type casting is changing a value that is of one data type into another, such as converting a string into an integer.

```python
value = "56"
print(type(value))

value = int(value)
print(type(value))
```

Notice on the first line, we created a string variable that contained the number 56 within it. When we print out the type of that value, you will see that is a string. We then use the `int()` function to convert whatever is in the parenthesis, in this case the original value, into an integer. Keep in mind that if your string is not a number, this will not work, and instead return an error.

```python
s = "Hello"
s = int(s)
print(s)
```

As you can see, we get an error because the string `"Hello"` cannot be converted into an integer because it is not a number. One way we can check if a string is a number is by using the `isdigit()` function on a string

```python
print("56".isdigit())
print("Hello".isdigit())
```

Using the other data types that we listed previously, we can convert them around one another. Let's take a look at some of these interactions.

```python
# Lets define some values

val_1 = "Hello"
val_2 = 83
val_3 = 922.1
val_4 = True

print(type(val_1), type(val_2), type(val_3), type(val_4))

# Now let's do some type casting
val_1 = bool(val_1)
val_2 = float(val_2)
val_3 = int(val_3)
val_4 = str(val_4)

# \n is known as the new-line character, allows you to create multiline print statements.

print(f'New Values: \n{val_1}\n{val_2}\n{val_3}\n{val_4}')
print(type(val_1), type(val_2), type(val_3), type(val_4))

```

As you can see, variables can be casted in a multitude of different ways.
